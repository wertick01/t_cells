```python
import json
from os import listdir
from os.path import isfile, join

gr_dct = {}
pr_dct = {}

for i in [f for f in listdir('/home/mrred/Рабочий стол/Работа/t_cells/Allign/Groups/') if isfile(join('/home/mrred/Рабочий стол/Работа/t_cells/Allign/Groups/', f))]:
    with open(f'/home/mrred/Рабочий стол/Работа/t_cells/Allign/Groups/{i}', 'r', encoding='utf-8') as ft:
        group = json.load(ft)
        gr_dct[i] = group

for i in [f for f in listdir('/home/mrred/Рабочий стол/Работа/t_cells/Allign/Primers/') if isfile(join('/home/mrred/Рабочий стол/Работа/t_cells/Allign/Primers/', f))]:
    with open(f'/home/mrred/Рабочий стол/Работа/t_cells/Allign/Primers/{i}', 'r', encoding='utf-8') as ft:
        primer = json.load(ft)
        pr_dct[i] = primer
```

```python
final_lst = []

for group_name, group_value in gr_dct.items():
    #print(group_name, group_value)
    for gr_val in group_value:
        #print(group_name, gr_val['key'], gr_val['value'])
        for primer_group, primers in pr_dct.items():
            for primer in primers:
                gr_val['value'] = gr_val['value'].upper()
                rat = fuzz.partial_ratio(gr_val['value'], primer['primer'])
                if rat >= 50 and len(gr_val['value']) > 300:
                    final_lst.append(
                        {'primer_group': primer_group, 
                        'primer_key': primer['key'], 
                        'primer': primer['primer'], 
                        'sequence_group': group_name, 
                        'sequence_key': gr_val['key'], 
                        'sequence': gr_val['value'], 
                        'ratio': rat})
```

```python
def longestSubstringFinder(string1, string2):
    answer = ""
    len1, len2 = len(string1), len(string2)
    for i in range(len1):
        match = ""
        for j in range(len2):
            if (i + j < len1 and string1[i + j] == string2[j]):
                match += string2[j]
            else:
                if (len(match) > len(answer)): answer = match
                match = ""
    return answer

longestSubstringFinder(seq1, seq2)
```
